#### Q1 [3 points]

***Given a 2d numpy array, the task is to flatten a 2d numpy array into a 1d array.***

```
[[1, 2, 3], 
[2, 4, 5], 
[6, 8, 3]]
```

````python
## Soln 1

#reshape
import numpy as np
# Create 2D array
ini_arr1 = np.array([[1,2,3],[2,4,5],[6,8,3]])
print("2D ARRAY :\n",str(ini_arr1))
print("\n")
# Flatten it
res = ini_arr1.reshape([1,9])
# Print the output along with the dimension of flatten array
print("1D ARRAY (reshape) :\n", res)
print("\n")

#flatten
ini_arr1 = np.array([[1,2,3],[2,4,5],[6,8,3]])
print("2D ARRAY :\n",str(ini_arr1))
print("\n")
res = ini_arr1.flatten()
print("1D ARRAY (flatten) :\n", res)
print("\n")

#ravel
ini_arr1 = np.array([[1,2,3],[2,4,5],[6,8,3]])
print("2D ARRAY :\n",str(ini_arr1))
print("\n")
res = ini_arr1.ravel()
print("1D ARRAY (ravel) :\n", res)
````

#### Q2 [3 points]

***Age & Height of 20 students in two different arrays with age and height (in cms).*** 
```

height = [156,144,180,162,152,157,154,155,151,150,158,179,126,182,183,154,159,160,172,149]
```
***Print the heights and ages of those students whose height is above 156 cm. Print your answer in numpy array.***



````python
## Soln 2
# Data provided in question
import numpy as np
age = np.array([15,17,19,20,14,21,16,19,13,20,22,23,21,16,18,19,20,15,17,18])
height = np.array([156,144,180,162,152,157,154,155,151,150,158,179,126,182,183,154,159,160,172,149])

filtered_age_list = []
filtered_height_list = []

# Code below to print the height and age of students whose height is more than 156 cm

for i in range(min(len(age), len(height))):
  if height[i] > 156 :
    filtered_height_list.append(height[i])
    filtered_age_list.append(age[i])

filtered_age = np.array(filtered_age_list)
filtered_height = np.array(filtered_height_list)

print(f'Ages of students above 156 cm height : {filtered_age}')
print(f'Students height above 156 cm : {filtered_height}')
````

#### Q6 [10 Points]

***Create a whirlpool pattern using numpy arrays***  

**1) Understand what is a whirlpool -**
- *Whirpool starts from 0 at the centre of an array*
- *Every layer of whirlpool is incremented by 1*
- *The last layer of whirlpool can have any number but only between 1 to 10*

**2) Example to visualize the whirlpool whose last layer is filled with 3s -**
- *Whirpool starts from 0 at the center*
- *The first layer is filled with 1s*
- *The second layer is filled with 2s*
- *The third/last layer is filled with 3s*

**Below you can see the layers of whirlpool for the above example -**        
1.png

**Here is the reason why we named it a whirlpool pattern -**       
2.png

**3) Requirments of this question -**        
- *Create a function that generates such whirpool patterns given the digits to be used in last layer*
- *The last layer of whirpool should only allow numbers between 1 to 10 (inclusive).*
- *No hardcoding is allowed!* 


````python
## Soln 6
# Function to create whirlpool patterns
# Call the function
import numpy as np
def whirlpool():
  n = int(input("Enter the number: "))
  c = 0
  size = 2 * n -1
  cx = cy = n - 1
  lst = []
  for y in range(0,size+2):
    lst.append([])
    for x in range(0,size+2):
      lst[y].append(c + max(abs(cx-x+1),abs(cy-y+1)))
      ar = np.array(lst)
  print(ar)

whirlpool()
````

***For example:-***
- Raw data:- Here are scores of some students after 1 week in AI4LR batch.         
``` 
Harshini - 60, Pranav - 54, Danish - 0, Siddhi - 50, Lakshmi - 58,
Bhavana - 47, Aditya - 59, Dhakshna - 0, Monica - 52
```

- Create a dictionary of this data.
- Call your function on this data such that it displays top 3 performers along with respective scores.
- A similar output is expected as shown below:-           
<img src = "https://drive.google.com/uc?id=1EZOgMdsJmaTs_ZNMMQZ2bqu-nA0pOCZm" height = 200>
- We will be using some other customised data to check the functionality of your function.

````python
## Soln 7 
data = {'Name': ['Harshini','Pranav','Danish','Siddhi','Lakshmi','Bhavana','Aditya','Dhakshna','Monica'], 'Score': [60,54,0,50,58,47,59,0,52]}
import pandas as pd
data1 = pd.DataFrame(data)
# Code the function
def top():
  n=int(input("Enter number to display top best performers:"))
  d2=data1.nlargest(n, columns=['Score'], keep='all')
  d3=pd.DataFrame(d2)
  d3=d2.reset_index(drop=True)
  d3.index = d3.index + 1
  d3.index.name = 'Rank'
  print(f'|        TOP {n}       |')
  print("""----------------------""")
  display(d3)

def bad():
  n=int(input("Enter number to display top bad performers:"))
  d2=data1.nsmallest(n, columns=['Score'], keep='all')
  d3=pd.DataFrame(d2)
  d3=d2.reset_index(drop=True)
  d3.index = d3.index + 1
  d3.index.name = 'Worst Rank'
  print(f'|        WORST {n}       |')
  print("""----------------------""")
  display(d3)
# Call the function on dict data to fetch top 3 performers
K = int(input("Enter 1 to see Top best performers or Enter 2 to see top Bad performers:"))
if K == 1 :
  top()
elif K == 2 :
  bad()
else :
  print("Either Type 1 or 2")
````

````python
## Soln 10

# Read the image and convert it into an array
import cv2
import matplotlib.pyplot as plt
from PIL import Image
img=cv2.imread("/content/original_image.jpg")
image = Image.open('/content/original_image.jpg')
from numpy import asarray
data = asarray(image)

# Get the grayscale image
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Invert the gray image
inverted_gray_image = 255 - gray_image

# Blur the inverted image
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21),0)

# Invert the blur image
inverted_blurred_img = 255 - blurred_img

# Blend (divide) gray and inverted blur images - Pencil Sketch is ready
pencil_sketch_IMG = cv2.divide(gray_image, inverted_blurred_img, scale = 256.0)

# Display the original and Pencil sketch images using suplots
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original Image', size=18,color="Red")
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Pencil Sketch of the Original Image', size=18)
nps = cv2.cvtColor(pencil_sketch_IMG, cv2.COLOR_BGR2RGB)
plt.imshow(nps)
plt.axis('off')
plt.show()

#The above code works perfectly, just copy and paste the required image path in code and see the result.
#I runned code uisng google colab path also, and result shown below. 
# the same code can run in kaggle using kaggle dataset path. 

````
